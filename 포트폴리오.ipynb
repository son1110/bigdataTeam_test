```python
# 각 행정구역별 20~29세 인구 수
import csv
f = open('대전 연령별인구현황.csv')
data = csv.reader(f)
next(data)
next(data)
result = []
for row in data :
    row[1:] = map(int, list(map(lambda x : x.replace(',',''), row[1:])))
    result.append(row[5])
    
print(len(result))
print(result)
```

```python
import csv
import matplotlib.pyplot as plt

with open('대전 연령별인구현황.csv') as f:
    data = csv.reader(f)
    next(data)
    next(data)
    result = []
    for row in data:
        row[1:] = map(int, list(map(lambda x : x.replace(',',''), row[1:])))
        result.append(row[5])

plt.rc('font', family = 'Malgun Gothic')        
plt.title('대전 행정구역별 20~29세 인구 수')
plt.bar(range(len(result)), result)
plt.show()
# 0:동구 1:중구 2:서구 3:유성구 4:대덕구
```

```python
import pandas as pd
```

```python
data = pd.read_csv('대전광역시_대학교 현황.csv', encoding='cp949')
data.head()
```

```python
data.isnull().sum()
```

```python
data['국립대'] = data['국립대'].fillna(0)
data['사립대'] = data['사립대'].fillna(0)
```

```python
data = data.dropna(subset=['연번'])
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
center = [36.350491, 127.384824]  
m=folium.Map(location=center, zoom_start=12)
```

```python
import numpy as np

data['total'] = np.where(data['국립대']==0, data['사립대'], data['국립대'])
data
```

```python
location_data=data[['위도','경도']].values[:len(data)].tolist()
location_data
```

```python
center = [36.350491, 127.384824]  
m=folium.Map(location=center, zoom_start=12)


for i in range(len(data)):   
    college_id=data['기관명'][i]
    total_count=int(data['구분'][i])   
    if data['구분'][i] == 1 : # 1 : 국립
        college_color='blue'
    elif data['구분'][i] == 2 : # 2 : 사립
        college_color='red'
    else :college_color='black'    
    folium.Circle(location=location_data[i],radius=total_count,color=college_color,fill=True,fill_opacity=1,tooltip=college_id,popup=total_count).add_to(m)

    
m 
```

```python
from folium import plugins 

center = [36.350491, 127.384824]  
m=folium.Map(location=center, zoom_start=11)


for i in range(len(data)):   
    college_id=data['기관명'][i]
    total_count=int(data['구분'][i])   
    if data['구분'][i] == 1 : # 1 : 국립
        college_color='blue'
    elif data['구분'][i] == 2 : # 2 : 사립
        college_color='red'
    else :college_color='black'    
    folium.Circle(location=location_data[i],radius=total_count,color=college_color,fill=True,fill_opacity=1,tooltip=college_id,popup=total_count).add_to(m)

plugins.MarkerCluster(location_data).add_to(m)

m 
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
size = [7, 2, 5, 6, 0]
label = ['동구','중구','서구', '유성구', '대덕구']
color = ['bisque', 'lightsteelblue', 'thistle', 'lavender', 'lightgrey']
plt.axis('equal')
plt.pie(size, labels = label, autopct = '%.1f%%', colors = color)
plt.title('대전 행정구역별 대학교 비율')
plt.legend()
plt.show()
```

```python
import csv
f = open('대전 연령별인구현황 - 행렬바꿈.csv')
data = csv.reader(f)
next(data)
next(data)
result = []
for row in data :
    row[1:] = map(int, list(map(lambda x : x.replace(',',''), row[1:])))
    result.append(row[1])
    
print(len(result))
print(result)
```

```python
import csv
import matplotlib.pyplot as plt

with open('대전 연령별인구현황 - 행렬바꿈.csv') as f:
    data = csv.reader(f)
    next(data)
    next(data)
    result = []
    for row in data:
        row[1:] = map(int, list(map(lambda x : x.replace(',',''), row[1:])))
        result.append(row[1])

plt.rc('font', family = 'Malgun Gothic')        
plt.title('동구 연령별 인구수')
plt.bar(range(len(result)), result)
plt.show()
# 0: 0~9세 1: 10~19세 2: 20~29세 3: 30~39세 4: 40~49세 5: 50~59세
# 6: 60~69세 7: 70~79세 8: 80~89세 9: 90~99세 10: 100세 이상
```

#### ★결론 : 동구는 서구,유성구에 비해 인구 비율이 적다. 그런데 대학가 근처에는 20대의 분포가 높지만 이와 동시에 40-60대의 분포도 높기 때문. 그리고 대학가를 벗어나는 동구의 행정구역은 40-60대의 분포가 월등히 높다.
